```python
X = df.drop("price_range", axis=1)
y = df["price_range"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
loj_model = LogisticRegression(solver="liblinear").fit(X_train,y_train)
y_pred = loj_model.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier().fit(X_train,y_train)
y_pred = knn_model.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier().fit(X_train,y_train)
rf_model
y_pred = rf_model.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
models = [
    loj_model,
    knn_model,
    rf_model,
   
]

results = []
results = pd.DataFrame(columns=["Models","Accuracy"])

for model in models:
    Name = model.__class__.__name__
    y_pred = model.predict(X_test)
    Accuracy = accuracy_score(y_test,y_pred)
    result = pd.DataFrame([[Name, Accuracy*100]], columns = ["Models","Accuracy"])
    results = results.append(result)
```

```python
results
```

```python

from sklearn.model_selection import cross_val_score

scores3 = cross_val_score(loj_model, X, y, cv=10)
print('Cross-Validation Accuracy Scores', scores3)

```

```python
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores3.mean(), scores3.std()))
print("Accuracy: {:.1f} %".format(scores3.mean()*100))
```

```python
from sklearn.model_selection import cross_val_score

scores1 = cross_val_score(knn_model, X, y, cv=10)
print('Cross-Validation Accuracy Scores', scores1)
```

```python
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores1.mean(), scores1.std()))
print("Accuracy: {:.1f} %".format(scores1.mean()*100))
```

```python
from sklearn.model_selection import cross_val_score

scores2 = cross_val_score(rf_model, X, y, cv=10)
print('Cross-Validation Accuracy Scores', scores2)
```

```python
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores2.mean(), scores2.std()))
print("Accuracy: {:.1f} %".format(scores2.mean()*100))
```

```python

    
sco = [
    scores3,
    scores1,
    scores2,
   
]

_results = []
_results = pd.DataFrame(columns=["AccuracyCrossVal"])

for s in sco:
    Name = s
    Accuracy1 = s.mean()*100
    _result = pd.DataFrame([[Accuracy1]], columns = ["AccuracyCrossVal"])
    _results = _results.append(_result)
_results
```
